```python

df = pd.read_csv('MarijuanaLegalization_byState.csv')
df
```

```python
modDf = df.drop([df.index[51] , df.index[52], df.index[53], df.index[54], df.index[55]])
del modDf['State Laws']
modDf
```

```python
#'0000' string if state has not voted in Medicinal use
modDf['Year-Medicinal'] = modDf['Year-Medicinal'].replace(np.nan, '0000')
#'0000' string if state has not voted in Recreational use
modDf['Year-Recreational'] = modDf['Year-Recreational'].replace(np.nan, '0000')
#Replace Nan's with empty strings
modDf['Comments'] = modDf['Comments'].replace(np.nan, '')
modDf
```

```python
#Function to assign cateogory #'s depending on Legalization status per state
def func(var):
    if var == 'Fully Illegal':
        return 0
    elif var =='Mixed':
        return 1
    elif var =='Fully Legal':
        return 2
    else:
        return 3

#List Comprehension: Creating a category column and adding it the datafram after the self-defined function
modDf['Legalization Category'] = [func(val) for val in modDf['Legal Status']]
modDf
```

```python
#Scrape website for the Lats and Longs for each state
dfLocations = pd.read_html('https://inkplant.com/code/state-latitudes-longitudes')
dfLocations
```

```python
#Concated the legalization df with the Latitude/Longitude list I scraped from a website with the info in a table
result = pd.concat([modDf, dfLocations[0]['Latitude'], dfLocations[0]['Longitude']], axis=1)
result
```

```python
# Store latitude and longitude in locations
locations = result[["Latitude", "Longitude"]]

# Fill NaN values and convert to float
rating = result["Legalization Category"].astype(float)

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=rating*15, 
                                 dissipating=False, max_intensity=25,
                                 point_radius=2)
# Add layer
fig.add_layer(heat_layer)
# Display figure
fig
```
